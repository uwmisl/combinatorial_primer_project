# Plotting primer behavior
Plotting the number of files each primer design can encode

tomek = n^2 - 3n + 2

organick = n / 2 

winston = (n / 2)^2

```python
import seaborn as sns # for plotting
import pandas as pd # organize data for seaborn plotting
import math # for rounding numbers
import matplotlib.pyplot as plt # for saving figure
```

```python
primer_data = {'Number of Primers':[], 'Primer Design': [], 'Max Number of Files': []} 
designs = ['Traditional','Nested Combinatorial', 'This Work']

for i in range(100):
    tomek = math.trunc(i**2 - 3*i + 2)
    organick = math.trunc(i / 2)
    winston = math.trunc((i / 2)**2)
    design_vals = [organick, tomek, winston]

    if i < 2:
        design_vals = [0,0,0]
        
    for j in range(len(designs)):
        primer_data['Number of Primers'].append(i)
        primer_data['Primer Design'].append(designs[j])
        primer_data['Max Number of Files'].append(design_vals[j])
           
primer_data = pd.DataFrame(primer_data)

# plotting
sns.set(context='paper', style='white', font_scale=1.8, rc={"figure.figsize":(15.5, 6)})
sns.lineplot(data=primer_data, y="Max Number of Files", x="Number of Primers", 
             hue='Primer Design').set(title='Comparing Primer Design Schemes')
plt.legend(title='Primer Design*')
plt.tight_layout()
plt.savefig('primer_behavior_plot.png', dpi=500)
```
